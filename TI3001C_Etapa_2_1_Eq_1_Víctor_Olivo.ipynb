# Base De Datos #2 Productos

```python
from google.colab import files
import pandas as pd

# Upload the file
uploaded = files.upload()

# Read the uploaded file into a pandas DataFrame
BaseDatosProductos = pd.read_excel('Productos.xlsx')

# Display the first few rows of the DataFrame
BaseDatosProductos.head()
```

```python
# Número de renglones
num_rows = len(BaseDatosProductos)

# Número de columnas
num_cols = len(BaseDatosProductos.columns)

# Tipo de variables y total de datos
data_types = BaseDatosProductos.dtypes
total_data = num_rows * num_cols

# Verificar el tipo de datos utilizando Pandas (muestra información detallada)
print("Información general de la base de datos:")
print(BaseDatosProductos.info())

print("\nResumen del análisis:")
print(f"Número de renglones: {num_rows}")
print(f"Número de columnas: {num_cols}")
print(f"Tipos de datos:\n{data_types}")
print(f"Total de datos: {total_data}")
```

*Se verificó la estructura de la base de datos y todas las variables
tienen el tipo de dato correcto. No es necesario realizar modificaciones.*

```python
missing_values = BaseDatosProductos.isnull().any()
print("Columns with missing values:\n", missing_values)
```

```python
df = BaseDatosProductos
nulos_por_columna = df.isnull().sum()
print("Cantidad de valores nulos por variable:")
print(nulos_por_columna)
```

```python
# Categorias
columnas_categoricas = BaseDatosProductos.select_dtypes(include=['object', 'category']).columns
print("\nCategorias:")
for col in columnas_categoricas:
    print("\n Frecuencia de cada categoría:")
    print(BaseDatosProductos[col].value_counts())
```

```python
# Rangos (variables numéricas)
columnas_numericas = BaseDatosProductos.select_dtypes(include=['int64', 'float64']).columns

for col in columnas_numericas:
  rango_max = BaseDatosProductos[col].max()
  rango_min = BaseDatosProductos[col].min()
  print(f"\n {col}: Rango = {rango_max} - {rango_min}")
  print(f"Rango = {rango_max - rango_min}")
```
